```python
# Import dataset
import pandas as pd
import numpy as np

df = pd.read_csv('test.csv')
pd.set_option("display.max_rows", df.shape[0] + 1)

display(df.head(10))
print("Jumlah data :", len(df))
```

```python
# Perbandingan quantity data
grouped = df.groupby('Sleep Disorder').count()

display(grouped.iloc[:, 0:1])
```

```python
# Pengecekan Missing Value
data_null = df.isnull().sum()
display(data_null)
```

```python
# Binerisasi Data
# Fitur Gender dirubah menjadi nilai 0 untuk female dan 1 untuk male
df['Gender'] = df['Gender'].replace('Male', 1)
df['Gender'] = df['Gender'].replace('Female', 0)

# Fitur BMI Categori dirubah menjadi nilai 0 untuk Normal dan Normal Weight, 1 untuk Overweight dan 2 untuk Obese
value_BMI = df['BMI Category'].unique()
binerisasi_BMI = {'Normal': 0, 'Normal Weight': 0, 'Overweight': 1, 'Obese' : 2}
df['BMI Category'] = df['BMI Category'].map(binerisasi_BMI)

# Fitur Sleep Disorder dirubah menjadi nilai 0 untuk None, 1 untuk Insomnia dan 2 untuk SLeep Apnea
value_sleep_disorder = df['Sleep Disorder'].unique()
binerisasi_sleep_disorder = {'None': 0, 'Insomnia': 1, 'Sleep Apnea': 2}
df['Sleep Disorder'] = df['Sleep Disorder'].map(binerisasi_sleep_disorder)

# Fitur Occupation dirubah menjadi nilai 0 untuk Software Engineer, 1 untuk Doctor 
# dan seterusnya sesuai dengan data binerisasi_occupation
value_occupation = df['Occupation'].unique()
binerisasi_occupation = {'Software Engineer' : 0, 'Doctor' : 1, 'Sales Representative' : 2, 'Teacher' : 3,
       'Nurse' : 4 , 'Engineer' : 5, 'Accountant' : 6, 'Scientist' : 7, 'Lawyer' : 8,
       'Salesperson' : 9, 'Manager' : 10}
df['Occupation'] = df['Occupation'].map(binerisasi_occupation)

# Pemecahan 1 fitur yaitu Blood Pressure menjadi 2 fitur yaitu Systolic dan Diastolic 
# untuk memudahkan proses penghitungan
for i in range(len(df)):
    blood_pressure = df['Blood Pressure'][i].split('/')
    df.loc[i, 'Systolic'] = blood_pressure[0]
    df.loc[i, 'Diastolic'] = blood_pressure[1]

# Pembuatan fitur baru
new_columns = ['Gender', 'Age', 'Occupation', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'BMI Category', 'Systolic', 'Diastolic', 'Heart Rate', 'Daily Steps', 'Sleep Disorder']
    
data = pd.DataFrame(df, columns=new_columns)
# display(data)
data.head()
```

```python
# Penentuan Variable independen dan dependen
# Variable independen
x = data.drop(['Sleep Disorder'], axis = 1)
x.head()

# Variable dependen
y = data['Sleep Disorder']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 314)
```

```python
from sklearn.naive_bayes import GaussianNB
# Fungsi klasifikasi Naive Bayes
modelnb = GaussianNB()

# Memasukan data training pada fungsi 
nbtrain = modelnb.fit(x_train, y_train)
```

```python
# Menentukan hasil prediksi dari x_test
y_pred = nbtrain.predict(x_test)
y_pred
```

```python
np.array(y_test)
```

```python
# Menentukan probabilitas hasil prediksi 
nbtrain.predict_proba(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
